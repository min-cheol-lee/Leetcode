```python
class Solution:
    def isNumber(self, s: str) -> bool:

        Scnt = 0
        Ncnt = 0
        Ecnt = 0
        Dcnt = 0
        
        for i, l in enumerate(s):    
            if not l in ["+","-","e","E","."]:
                if not l.isnumeric():
                    return ""
                elif s[i-1] in ["+","-"] and Ncnt==1 and Ecnt==0:
                    return ""
                else:
                    Ncnt = 1
                    continue
                 
            elif l=="+":
                if i==0:
                    continue
                elif Scnt==1:
                    if s[i-1]=="e" or s[i-1]=="E":
                        continue
                    else:
                        return ""
                elif Ecnt==1 and not s[i-1]=="e":
                    return ""
                elif s[i-1]==".":
                    return ""
                elif i==len(s)-1:
                    return ""
                else:
                    Scnt = 1
                    
            elif l=="-":
                if i==0:
                    continue
                elif Scnt==1:
                    if s[i-1]=="e" or s[i-1]=="E":
                        continue
                    else:
                        return ""
                elif Ecnt==1 and not s[i-1]=="e":
                    return ""
                elif s[i-1]==".":
                    return ""
                elif i==len(s)-1:
                    return ""
                else:
                    Scnt = 1
            
            elif l=="e" or l=="E":
                if Ecnt==1 or Ncnt==0 or i==len(s)-1:
                    return ""
                elif s[i-1] in ["+","-"]:
                    return ""
                else:
                    Ecnt = 1
                    continue
            
            elif l=='.':
                if Ecnt==1 or Dcnt==1:
                    return ""
                elif s[i-1] in ["+","-"] and Ncnt==1:
                    return ""
                else:
                    Dcnt = 1
                    continue
        
        if Ncnt == 1:
            return "true"
        else:
            return ""
                
                
```
